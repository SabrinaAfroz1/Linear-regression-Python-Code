```python
df=pd.read_csv("convertnumber.csv")
df.head(157)
```

```python
colums = (df.columns[0])
```

```python
max= [df[c].max() for c in df.columns] 
min= [df[c].min() for c in df.columns]
```

```python
i=0
for c in df.columns:
    while(i<len(df.columns)): 
        df[c]=((df[c]-min[i])/(max[i]-min[i]))
        i=i+1
        break
```

```python
arr = df.values
x_train=[]
y1=[]
a=df.shape
for i in range(a[0]):                      
    x_train.append((arr[i][:-1]).tolist())
    y1.append(arr[i][-1])
```

```python
m=np.ones((155,1))
b=np.matrix(x_train)
b=np.concatenate((m,b),axis=1)     
d=b.T
e=np.linalg.inv(np.matmul(d,b))
y1=np.matrix(y1)
y1=y1.T
f=np.matmul(e,d)
c1=np.matmul(f,y1)
```
