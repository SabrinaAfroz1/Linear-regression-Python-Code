```python
df=pd.read_csv("convertnumber.csv")
df.head(157)
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['age','sex','steroid','antivirals','fatigue','malaise','anorexia','liver_big','liver_firm','spleen_palpable','spiders','ascites','varices','bilirubin','alk_phosphate','sgot','albumin','protime','histology']], df.className)
```
